# Andamento campagna vaccinale anti Covid-19

Uso questo notebook per analizzare dati tratti da https://github.com/italia/covid19-opendata-vaccini per rispondere ad alcune domande che a volte mi pongo sulla campagna vaccinale anti Covid-19.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import  re, json, requests
```

## Upload dei dati

Data e ora di ultimo aggiornamento del dataset.

```python
path_update='https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/last-update-dataset.json'

resp = requests.get(path_update)
update = json.loads(resp.text)
time=update['ultimo_aggiornamento']

print(update)
```

"consegne-vaccini-latest.csv" contiene i dati sul totale delle consegne giornaliere dei vaccini suddivise per regioni.

```python
path_consegne='https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/consegne-vaccini-latest.csv'

df_consegne=pd.read_csv(path_consegne, error_bad_lines=False)
df_consegne.head()
```

"vaccini-summary-latest.csv" contiene i dati sul totale delle consegne e somministrazioni avvenute sino ad oggi, includendo la percentuale di dosi somministrate (sul totale delle dosi consegnate) suddivise per regioni.

```python
path_consegne_summ='https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/vaccini-summary-latest.csv'

df_consegne_summ=pd.read_csv(path_consegne_summ)
df_consegne_summ.head()
```

"somministrazioni-vaccini-latest.csv" contiene i dati sulle somministrazioni giornaliere dei vaccini suddivisi per regioni, fasce d'età e categorie di appartenenza dei soggetti vaccinati.

```python
path_somministrazioni='https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/somministrazioni-vaccini-latest.csv'

df_somministrazioni = pd.read_csv(path_somministrazioni, error_bad_lines=False)
df_somministrazioni.head()
```

"somministrazioni-vaccini-summary-latest.csv" contiene i dati sul totale delle somministrazioni giornaliere per regioni e categorie di appartenenza dei soggetti vaccinati.

```python
path_somministrazioni_summ='https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/somministrazioni-vaccini-summary-latest.csv'

df_somministrazioni_summ=pd.read_csv(path_somministrazioni_summ)
df_somministrazioni_summ.head()
```

"punti-somministrazione-latest" contiene i punti di somministrazione per ciascuna Regione e Provincia Autonoma.

```python
path_punti='https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/punti-somministrazione-latest.csv'

df_punti=pd.read_csv(path_punti)
df_punti.head()
```

"anagrafica-vaccini-summary-latest.csv" contiene i totali delle somministrazioni per fasce d'età.

```python
path_anagrafica_summ='https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/anagrafica-vaccini-summary-latest.csv'

df_anagrafica_summ=pd.read_csv(path_anagrafica_summ)
df_anagrafica_summ.head(10)
```

## Andamento per fasce d'età

```python
fasce=df_anagrafica_summ['fascia_anagrafica'].to_list()
prima_summ_fas=df_anagrafica_summ['prima_dose'].to_list()
seconda_summ_fas=df_anagrafica_summ['seconda_dose'].to_list()

# set width of bars
barWidth = 0.25

# Set position of bar on X axis
r1 = np.arange(len(prima_summ_fas))
r2 = [x + barWidth for x in r1]

# Make the plot
plt.bar(r1, prima_summ_fas, color='#FF7F50', width=barWidth, edgecolor='white', label='dose 1')
plt.bar(r2, seconda_summ_fas, color='#40E0D0', width=barWidth, edgecolor='white', label='dose 2')

plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)

# Add xticks on the middle of the group bars
plt.xlabel('Vaccinati per fascia di età', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(fasce))], fasce)

# Create legend & Show graphic
plt.legend()
plt.show()
```

```python
days=df_somministrazioni['data_somministrazione'].drop_duplicates().to_list()
nd=len(days) #numero di giorno da cui è iniziata la campagna di vaccinazione

categorie=df_somministrazioni.keys().to_list()[6:13]

print(categorie)
```

### Tutti i vaccini

```python
cat0=np.zeros(nd)
cat1=np.zeros(nd)
cat2=np.zeros(nd)
cat3=np.zeros(nd)
cat4=np.zeros(nd)
cat5=np.zeros(nd)
cat6=np.zeros(nd)



for i in range(len(days)):
    cat0[i]=df_somministrazioni[df_somministrazioni['data_somministrazione']==days[i]][categorie[0]].sum()
    cat1[i]=df_somministrazioni[df_somministrazioni['data_somministrazione']==days[i]][categorie[1]].sum()
    cat2[i]=df_somministrazioni[df_somministrazioni['data_somministrazione']==days[i]][categorie[2]].sum()
    cat3[i]=df_somministrazioni[df_somministrazioni['data_somministrazione']==days[i]][categorie[3]].sum()
    cat4[i]=df_somministrazioni[df_somministrazioni['data_somministrazione']==days[i]][categorie[4]].sum()
    cat5[i]=df_somministrazioni[df_somministrazioni['data_somministrazione']==days[i]][categorie[5]].sum()
    cat6[i]=df_somministrazioni[df_somministrazioni['data_somministrazione']==days[i]][categorie[6]].sum()
    if i>0:
        cat0[i]=cat0[i]+cat0[i-1]
        cat1[i]=cat1[i]+cat1[i-1]
        cat2[i]=cat2[i]+cat2[i-1]
        cat3[i]=cat3[i]+cat3[i-1]
        cat4[i]=cat4[i]+cat4[i-1]
        cat5[i]=cat5[i]+cat5[i-1]
        cat6[i]=cat6[i]+cat6[i-1]
```

```python
plt.plot( range(nd), cat0, marker='', color='blue', label=categorie[0])
plt.plot( range(nd), cat1, marker='', color='red', label=categorie[1])
plt.plot( range(nd), cat2, marker='', color='green', label=categorie[2])
plt.plot( range(nd), cat3, marker='', color='olive', label=categorie[3])
plt.plot( range(nd), cat4, marker='', color='maroon', label=categorie[4])
plt.plot( range(nd), cat5, marker='', color='purple', label=categorie[5])
plt.plot( range(nd), cat6, marker='', color='orange', label=categorie[6])


plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.legend()
plt.show()
```
